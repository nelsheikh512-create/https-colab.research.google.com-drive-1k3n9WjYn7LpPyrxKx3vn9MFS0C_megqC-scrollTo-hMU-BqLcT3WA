## EXERCISE 3: Add New Student (Fill 7 blanks)

**Challenge:** Validate age input AND add to dictionary

**Steps:**
1. Get name input
2. Get age input as string
3. **IF** age is valid digits ‚Üí convert to int
4. **IF** age > 0 ‚Üí add to database
5. **ELSE** ‚Üí show errors

````python
# EXERCISE 3 - FILL ALL 7 BLANKS:

name = input("Enter student name: ").strip()
age_str = input("Enter age: ").strip()

# BLANK 1-2: Check if age_str contains only digits
if age_str in():  # IF it's all digits?

    # BLANK 3: Convert string to integer
    age = int(age_str)

    # BLANK 4-5: Check if age is positive
    if age in 0:  # What about age greater than zero?

        # Get courses input
        courses_input = input("Enter courses (comma-separated): ").strip()
        courses_list = [course.strip() for course in courses_input.split(",")]

        # BLANK 6: Add to dictionary with nested structure
        student_db[name] = {"age": age, "courses": courses_list}

        print(f"‚úÖ Success! Added {name} to database!")

    else:  # ELSE age invalid
        print("‚ùå Error: Age must be greater than 0!")

else:  # age_str not digits
    print("‚ùå Error: Age must be a number!")

# Display updated database
print("\n=== UPDATED DATABASE ===")
for n in student_db:
    print(f"{n}: Age {student_db[n]['age']}, Courses: {', '.join(student_db[n]['courses'])}")
````

**Steps:**

````python
# EXERCISE 4 - FILL 5 BLANKS TO COMPLETE DELETE FUNCTIONALITY

# Student database
student_db = {
    "Amjad": {"age": 20, "courses": ["Python", "Math"]},
    "Yasmina": {"age": 22, "courses": ["Python", "Physics"]},
    "Salah": {"age": 33, "courses": ["Python", "Chemistry"]},
    "Hend": {"age": 30, "courses": ["Python", "Biology"]}
}

print("\n=== DELETE STUDENT ===")
name = input("Enter student name to delete: ").strip()

# BLANK1: Start IF condition to check if student exists
# BLANK2: Use membership operator to check if name is in dictionary
if name in student_db:

    # Display student information before deleting
    student = student_db[name]
    print(f"\nüìã Student to delete:")
    print(f"Name: {name}")
    print(f"Age: {student['age']}")
    print(f"Courses: {', '.join(student['courses'])}")

    # BLANK3: Convert user input to lowercase for case-insensitive comparison
    confirm = input("\n‚ö†Ô∏è Are you sure? (yes/no): ").strip().lower()

    # BLANK4: Check if user confirmed with 'yes'
    if confirm == "yes":
        # BLANK5: Use del statement to remove student from dictionary
        del student_db[name]
        print(f"\n‚úÖ Success! Deleted {name} from database!")
    else:
        print(f"\n‚ö†Ô∏è Deletion cancelled. {name} remains in database.")

else:
    print(f"\n‚ùå Error: Student '{name}' not found in database!")

# Display remaining students to verify deletion
print(f"\nüìä Total students remaining: {len(student_db)}")
print(f"Students: {', '.join(student_db.keys())}")
````
